```python
a = 1.105
b = 6.453
c = 3.54
```

```python
precision_power = lambda num: len(str(num).split(".")[1]) 
```

```python
d_a = 5*10**-(precision_power(a)+1)
d_b = 5*10**-(precision_power(b)+1)
d_c = 5*10**-(precision_power(c)+1)
```

```python
print(d_a)
print(d_b)
print(d_c)
```

```python
Lower_border = lambda x, d_x: x - d_x
Higher_border = lambda x, d_x: x + d_x
```

```python
l_a, l_b, l_c = Lower_border(a, d_a), Lower_border(b, d_b), Lower_border(c, d_c)
h_a, h_b, h_c = Higher_border(a, d_a), Higher_border(b, d_b), Higher_border(c, d_c)
```

```python
print(l_a, l_b, l_c)
print(h_a, h_b, h_c)
```

```python
l_b_minus_c = l_b - h_c
h_b_minus_c = h_b - l_c
```

```python
print(l_b_minus_c)
print(h_b_minus_c)
```

```python
import math
```

```python
def truncate(number, digits):
    stepper = 10.0 ** digits
    return math.trunc(stepper * number) / stepper
```

```python
l_b_minus_c = truncate(l_b_minus_c, min(precision_power(l_b), precision_power(h_c))+1)
h_b_minus_c = truncate(h_b_minus_c, min(precision_power(l_b), precision_power(h_c))+1)
```

```python
print(l_b_minus_c)
print(h_b_minus_c) # последнее число "дополнительное"
```

```python
l_square = l_b_minus_c**2
h_square = h_b_minus_c**2
```

```python
print(l_square, h_square)
```

```python
l_square = truncate(l_square, precision_power(l_b_minus_c))
h_square = truncate(h_square, precision_power(h_b_minus_c)) + 10**-(precision_power(h_b_minus_c))
```

```python
print(l_square, h_square)
```

```python
l_2a, h_2a = l_a*2, h_a*2
```

```python
print(l_2a, h_2a)
```

```python
l_2a_plus_b = l_2a + l_b
h_2a_plus_b = h_2a + h_b
```

```python
print(l_2a_plus_b, h_2a_plus_b)
```

```python
l_2a_plus_b = truncate(l_2a_plus_b, min(precision_power(l_2a), precision_power(l_b)))
h_2a_plus_b = truncate(h_2a_plus_b, min(precision_power(h_2a), precision_power(h_b))) + 10**-(min(precision_power(h_2a), precision_power(h_b)))
```

```python
print(l_2a_plus_b, h_2a_plus_b)
```

```python
l_z = l_square / h_2a_plus_b
h_z = h_square / l_2a_plus_b
```

```python
print(l_z, h_z)
```

```python
l_z = truncate(l_z, 3)
h_z = truncate(h_z, 3)
```

```python
print(l_z, h_z)
```
