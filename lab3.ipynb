```python
def gauss(*rows):
    matrix = []
    results = []
    for row in rows:
        current = []
        for i in range(0, len(row)-1):
            current.append(row[i])
        matrix.append(current)
        results.append(row[len(row)-1])
    for i in range(len(matrix)-1):
        if(matrix[i][i] == 0):
            for j in range(i+1, len(matrix)):
                if(matrix[j][i] != 0):
                    temp = matrix[i].copy()
                    temp_res = results[i]
                    results[i] = results[j]
                    matrix[i] = matrix[j]
                    results[j] = temp_res
                    matrix[j] = temp
                    break
        if matrix[i][i] != 0:
            for j in range(i+1, len(matrix)):
                d = matrix[j][i] / matrix[i][i]
                for k in range(i, len(matrix[j])):
                    matrix[j][k] = matrix[j][k]- matrix[i][k] * d
                results[j] = results[j] - results[i] * d
    solutions = [0] * len(matrix)
    for i in range(len(matrix)-1, -1, -1):
        if(matrix[i][i] == 0):
            continue
        solutions[i] = results[i]
        for j in range(i+1, len(matrix)):
            solutions[i] -= matrix[i][j] * solutions[j]
        solutions[i] /= matrix[i][i]
                    
    print(matrix)
    print(results)
    print(solutions)
    return solutions
```

```python
def inverse_matrix(*rows):
    matrix = []
    for row in rows:
        current = []
        for i in range(len(row)):
            current.append(row[i])
        matrix.append(current)
    
    n = len(matrix)

    inv_matrix = [[0.0] * n for _ in range(n)]
    for i in range(n):
        inv_matrix[i][i] = 1.0

    for i in range(n):
        if matrix[i][i] == 0:
            for j in range(i+1, n):
                if matrix[j][i] != 0:
                    matrix[i], matrix[j] = matrix[j], matrix[i]
                    inv_matrix[i], inv_matrix[j] = inv_matrix[j], inv_matrix[i]
                    break
        
        if matrix[i][i] == 0:
            raise ValueError("Матрица вырождена, обратной не существует")
        d = matrix[i][i]
        for k in range(n):
            matrix[i][k] /= d
            inv_matrix[i][k] /= d
        for j in range(n):
            if i != j:
                factor = matrix[j][i]
                for k in range(n):
                    matrix[j][k] -= matrix[i][k] * factor
                    inv_matrix[j][k] -= inv_matrix[i][k] * factor
    print("Обратная:")
    for row in inv_matrix:
        print([round(x, 4) for x in row])
    
    return inv_matrix
```

```python
def obratnaya(*rows):
    matrix = []
    results = []
    solutions = []
    for row in rows:
        current = []
        for i in range(0, len(row)-1):
            current.append(row[i])
        matrix.append(current)
        results.append(row[len(row)-1])
    inv = inverse_matrix(*matrix)
    for i in range(len(results)):
        solutions.append(sum(map(lambda x: x[0] * x[1], zip(inv[i], results))))
    return solutions
```

```python
def counting(A, B, x):
    new_x = []
    for i in range(len(A)):
        x_copy = x.copy()
        x_copy[i] = 0
        new_x.append((-sum([a * b for a, b in zip(x_copy, A[i])]) + B[i])/A[i][i])
    return new_x
```

```python
print(counting([[7, 4,-1],[2, 6, 3],[-1, 1, 4]], [7, -2, 4], [0,0,0]))
```

```python
def yakobi(x0: list[float], eps: float, *rows):
    matrix = []
    results = []
    for row in rows:
        current = []
        for i in range(0, len(row)-1):
            current.append(row[i])
        matrix.append(current) # A
        results.append(row[len(row)-1]) # B
    x_prev = x0
    x_new = counting(matrix, results, x_prev)
    while(max([abs(c-d) for c,d in zip(x_new, x_prev)]) >= eps):
        x_prev = x_new
        x_new = counting(matrix, results, x_prev)
    return x_new
```

```python
print(yakobi([0.7,1,2,0.5], 0.01, [3,1,-1,1,3],[1,-4,1,-1,-5],[-1,1, 4, 1, 14],[1,2,1,-5,3]))
```

```python
def counting_extended(A, B, x):
    new_x = []
    for i in range(len(A)):
        x_copy = x.copy()
        x_copy = new_x + x_copy[i:]
        x_copy[i] = 0
        new_x.append((-sum([a * b for a, b in zip(x_copy, A[i])]) + B[i])/A[i][i])
    return new_x
```

```python
def zeidel(x0: list[float], eps: float, *rows):
    matrix = []
    results = []
    for row in rows:
        current = []
        for i in range(0, len(row)-1):
            current.append(row[i])
        matrix.append(current) # A
        results.append(row[len(row)-1]) # B
    x_prev = x0
    x_new = counting_extended(matrix, results, x_prev)
    while(max([abs(c-d) for c,d in zip(x_new, x_prev)]) >= eps):
        x_prev = x_new
        x_new = counting_extended(matrix, results, x_prev)
    return x_new
```
