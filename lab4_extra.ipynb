```python
import numpy as np
import numdifftools as nd
from scipy.optimize import minimize_scalar
```

```python
def f(x, y):
    return x**2 - y

def g(x, y):
    return x + y - 6
```

```python
def Phi(xy):
    x, y = xy
    return f(x, y)**2 + g(x, y)**2
```

```python
def steepest_descent(x0, y0, max_iter=1000):
    x, y = x0, y0
    
    for k in range(max_iter):
        grad = nd.Gradient(Phi)([x, y])

        direction = -grad

        phi_alpha = lambda alpha, xc=x, yc=y, dc=direction: Phi(np.array([xc, yc]) + alpha * dc)
        
        result = minimize_scalar(phi_alpha, method='brent')
        alpha = result.x

        x_new = x + alpha * direction[0]
        y_new = y + alpha * direction[1]

        if Phi([x_new, y_new]) >= Phi([x, y]):
            break
        
        x, y = x_new, y_new
    
    return x, y
```

```python
x_sol, y_sol = steepest_descent(x0=0, y0=0)
print(f"\nРешение: x = {x_sol:.6f}, y = {y_sol:.6f}")
print(f"Проверка: f(x,y) = {f(x_sol, y_sol):.6f}, g(x,y) = {g(x_sol, y_sol):.6f}")
```
