```python
def jakobi(x0,eps, *rows):
    xs = [x0]
    matrix = []
    for row in rows:
        matrix.append(row)
    for index, row in enumerate(matrix):
        current = []
        for item_index, item in enumerate(row):
            new_item = 0 if index == item_index else (-1)*item/row[index]
            current.append(new_item)
        matrix[index] = current
    x1 = []
    for row in matrix:
        x1.append(sum(map(lambda x, y: x * y,row, x0)))
    xs.append(x1)
    while(max(map(lambda x, y: abs(x - y),xs[-1], xs[-2])) >= eps):
        x_new = []
        for row in matrix:
            x_new.append(sum(map(lambda x, y: x * y,row, xs[-1])))
        xs.append(x_new)
    return xs[-1]

print(jakobi([1,1], 0.005, [2,1], [1, 3]))
```
