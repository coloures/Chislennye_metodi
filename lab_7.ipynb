```python
import math
```

```python
func = lambda x: 2*x**2 - math.sqrt(x + 2)
```

```python
def rectangle_method(func, a, b, n, offset = "left"):
    h = (b - a)/n
    if (offset == "left"):
        quantities = [ func(a + h*i) for i in range(n)]
    else:
        quantities = [ func(a + h*i) for i in range(1, n+1)]
    return h * sum(quantities)
```

```python
rectangle_method(func, -2, 4, 6)
```

```python
rectangle_method(func, -2, 4, 6, "right")
```

```python
def trapezoid_method(func, a, b, n):
    h = (b - a)/n
    return h * ((func(a)+func(b))/2 + sum([func(a+h*i) for i in range(1, n)]))
```

```python
def parabola_method(func, a, b, n):
    h = (b - a)/n
    return h/3*(sum([func(a+h*2*i) + 4 * func(a+h*(2*i + 1)) + func(a+h*(2*i + 2)) for i in range(n // 2)]))
```
