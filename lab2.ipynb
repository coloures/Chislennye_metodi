```python
def dichotomy(f, a, b, eps):
    mid = (a+b)/2
    fa = f(a)
    fmid = f(mid)
    fb = f(b)
    if(fa*fb > 0):
        raise Exception()
    else:
        if(abs(fa-fb) < eps):
            return mid
        elif(fmid*fa < 0):
            return dichotomy(f, a, mid, eps)
        else:
            return dichotomy(f, mid, b, eps)
```

```python
dichotomy(lambda x: x**3+2*x+2,-1, 1, 0.01)
```

```python
import numdifftools as nd
import numpy as np
```

```python
def newton(f, a, b, eps):
    if(f(a)*nd.Derivative(f, n=2)(a) >0):
        current = a
        next = a - f(a)/nd.Derivative(f, n=1)(a)
    elif(f(b)*nd.Derivative(f, n=2)(b) >0):
        current = b
        next = b - f(b)/nd.Derivative(f, n=1)(b)
    else:
        raise Exception()
    while(abs(current)>eps and abs(next - current) > eps):
        current = +next
        next = current - f(current)/nd.Derivative(f, n=1)(current)
    return next
```

```python
newton(lambda x: x**3+2*x+2,-1, 1, 0.01)
```

```python
def simple_iter(f, x, eps):
    M = 1.01 * nd.Derivative(f, n = 1)(x)
    phi = lambda y: y - f(y)/ M
    current = x
    next = phi(x)
    while(abs(next - current) > eps):
        current = +next
        next = phi(current)
    return next
```

```python
simple_iter(lambda x: x**3+2*x+2,-1, 0.01)
```
