```python
f = lambda x: x**2 - 2 
```

```python
def hordes(a, b, f, eps):
    x0, x1 = a, b
    if(f(x1)*f(x0) > 0):
        raise ValueError("Нет в этом диапазоне нуля")
    while(abs(x1-x0) > eps and abs(f(x1)) > eps):
        temp = x0 - f(x0) / (f(x1) - f(x0))*(x1-x0)
        if(f(x0) * f(x1) < 0):
            x1 = temp
        else:
            x0 = temp
    return x0 - f(x0) / (f(x1) - f(x0))*(x1-x0)
```

```python
hordes(1, 2, f, 10**-6)
```

```python
def steffenson(x0, eps, f):
    x_current = x0
    while(abs(f(x_current)) > eps):
        func = f(x_current)
        x_current = x_current - (func**2) / (f(x_current + func) - func)
    return x_current
```

```python
steffenson(1, 10**(-6), f)
```
